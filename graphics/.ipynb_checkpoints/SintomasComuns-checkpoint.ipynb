#Análise de dados Impactos da Covid19 ES

```python
import pandas as pd
import matplotlib.pyplot as plt
covid19 = pd.read_csv("../datasets/covid-19-espirito-santo.csv", sep=";", encoding="ISO-8859-1")

colunasSintomas = ['Febre','DificuldadeRespiratoria','Tosse','Coriza','DorGarganta','Diarreia','Cefaleia']
covid19Confirmados = covid19.query('Classificacao == "Confirmados"')
covid19Sintomas = covid19Confirmados.filter(items=colunasSintomas);

totalFebre = len(covid19Sintomas[covid19Sintomas["Febre"] == "Sim"])
totalResp = len(covid19Sintomas[covid19Sintomas['DificuldadeRespiratoria'] == "Sim"])
totalTosse = len(covid19Sintomas[covid19Sintomas['Tosse'] == "Sim"])
totalCoriza = len(covid19Sintomas[covid19Sintomas['Coriza'] == "Sim"])
totalGarganta = len(covid19Sintomas[covid19Sintomas['DorGarganta'] == "Sim"])
totalDiarreia = len(covid19Sintomas[covid19Sintomas['Diarreia'] == "Sim"])
totalCefaleia = len(covid19Sintomas[covid19Sintomas['Cefaleia'] == "Sim"])
todosOsSintomas = [
    totalFebre, totalResp,
    totalTosse, totalCoriza,
    totalGarganta, totalDiarreia,
    totalCefaleia
]

totalSintomas = sum(todosOsSintomas)
percentualFebre = round((totalFebre / totalSintomas) * 100)
percentualResp = round((totalResp / totalSintomas) * 100)
percentualTosse = round((totalTosse / totalSintomas) * 100)
percentualCoriza = round((totalCoriza / totalSintomas) * 100)
percentualGarganta = round((totalGarganta / totalSintomas) * 100)
percentualDiarreia = round((totalDiarreia / totalSintomas) * 100)
percentualCefaleia = round((totalCefaleia / totalSintomas) * 100)

indexes = [
    'Febre (' + str(percentualFebre) + '%)',
    'Dificuldade Respiratória (' + str(percentualResp) + '%)',
    'Tosse (' + str(percentualTosse) + '%)',
    'Coriza (' + str(percentualCoriza) + '%)',
    'Dor de Garganta (' + str(percentualGarganta) + '%)',
    'Diarreia (' + str(percentualDiarreia) + '%)',
    'Cefaleia (' + str(percentualCefaleia) + '%)'
]

dataframe = pd.DataFrame({    
    'sintomas': todosOsSintomas
}, index=indexes)

dataframe.plot.pie(y='sintomas', figsize=(7, 7), legend=False)
plt.axis('off')
```
